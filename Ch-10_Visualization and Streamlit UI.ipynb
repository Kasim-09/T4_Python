```python
import numpy as np
```

```python
x = np.linspace(0,4*np.pi,1000)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("A sine curve")
plt.grid(True)
plt.show()
```

## SUBPLOTS

```python
plt.subplot(1,2,1) #1-row,2-col,1st figure
plt.plot(x,y,color="b")
plt.title("x vs y")

plt.subplot(1,2,2)
plt.plot(x,y,color="orange")
plt.title("y vs x")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,3,1) 
plt.plot(x,y,color="b")
plt.title("x vs y")

plt.subplot(2,3,2)
plt.plot(x,y,color="orange")
plt.title("y vs x")

plt.subplot(2,3,3) 
plt.plot(x,y,color="b")
plt.title("x vs y")

plt.subplot(2,3,4)
plt.plot(x,y,color="orange")
plt.title("y vs x")

plt.subplot(2,3,5) 
plt.plot(x,y,color="b")
plt.title("x vs y")

plt.subplot(2,3,6)
plt.plot(x,y,color="orange")
plt.title("y vs x")
plt.suptitle("A case of subplots")
plt.tight_layout()
plt.show()
```

## Scatter plot

```python
x = [1,2,3,4,5]
y = [10,14,20,45,60]
plt.scatter(
    x,y,
    color = 'm',
    s = 100,
    alpha = 0.9
)
plt.show()
```

## BAR PLOT VERTICAL 

```python
plt.bar(x,y,color = 'g',width = 0.5)
plt.show()
```

## BAR PLOT HORIZONTAL 

```python
plt.barh(x,y,color = 'cyan',height = 0.5)
plt.show()
```

## Pie Chart

```python
labels = ['A','B','C','D']
sizes = [40,35,10,15]
# colors = ['red','yellow','green','cyan']
```

```python
plt.pie(
    sizes,
    labels = labels,
    # colors = colors,
    explode = [0.1,0,0,0],
    startangle = 90,
    autopct = "%1.1f%%"
)
plt.show()
```

## Area plots

```python
import numpy as np 
import matplotlib.pyplot as plt

x = range(1,6)
y = [1,4,6,8,4]

plt.fill_between(x,y)
plt.plot(x,y,color='k',alpha=0.6)
plt.show()
```

## Stackplot 

```python
import numpy as np 
import matplotlib.pyplot as plt

x = range(1,6) #1,2,3,4,5
y1 = [1,4,6,8,9]
y2 = [2,2,7,10,12]
y3 = [2,8,5,10,6]

plt.stackplot(x,y1,y2,y3,labels = ['A','B','C'])
plt.legend()
plt.show()
```
