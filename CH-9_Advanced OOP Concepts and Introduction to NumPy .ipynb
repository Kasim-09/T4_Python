## INHERITANCE
- A class can inherit from other class
- Inheritance improves code reuse 
- Constructors, attributes, and , method are inherited from parent to child 
- The parent has no access to child class 
- child class can overrides method of parent class. It is called method overriding
- `super()` method is used to invoke parent class method and constructors

## `super()` method
- If we write constructor in both the parent and child class, the parent class constructor is not available to child class 
- `super()` method is used to invoke parent class constructor and method from child class

````python
class Father:
    def __init__(self):
        self.money = 10000
        print("Father class constructor")
    def show(self):
        print("Father class method")
        
class Son(Father):
    def __init__(self):
        self.money = 100
        self.car = 'BMW'
        super().__init__()
        print("Son class constructor")
    def show(self):
        super().show()
        print("Son class method")
````

````python
s1 = Son()
````

````python
s1.show()
````

````python
s1.money
````

````python
s1.money = 900
````

## WAP to create abstarct class Shape. It has abstract method calculate_area. Create derived classes Rectangle and circle and corresponding calculate area methods to find the area

````python
from abc import ABC, abstractmethod
from math import pi
class Shape(ABC):
       
    @abstractmethod
    def calculate_area(self):
         pass
````

````python
class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def calculate_area(self):
        return self.l * self.b
````

````python
class Circle(Shape):
    def __init__(self,r):
        self.r=r
    def calculate_area(self):
        return pi*self.r*self.r
````

## Operators Overloading
```
__gt__ -> >
__lt__ -> <
__eq__ -> ==
__add__ -> +
__sub__ -> -
__mul__ -> *
__truediv__ -> /

## Demonstraction through fraction class 
```
Fraction1 = 5/8
Fraction2 = 2/3
```

````python
class Fraction:
    def __init__(self,x,y):
        self.num = x
        self.den = y
    def __str__(self): # To print object we use str method 
        return f"{self.num}/{self.den}"
    def __add__(self,other):
        new_num = self.num * other.den + other.num*self.den
        new_den = self.den*other.den
        return f"{new_num}/{new_den}"
    def __sub__(self,other):
        new_num = self.num * other.den - other.num*self.den
        new_den = self.den*other.den
        return f"{new_num}/{new_den}"
    def __mul__(self,other):
        new_num = self.num * other.num 
        new_den = self.den*other.den
        return f"{new_num}/{new_den}"
    def __truediv__(self,other):
        new_num = self.num * other.den 
        new_den = self.den*other.num
        return f"{new_num}/{new_den}"
    def convert_decimal(self):
        return self.num / self.den
    def __eq__(self,other):
        return self.convert_decimal()==other.convert_decimal()
    def __lt__(self,other):
        return self.convert_decimal()<other.convert_decimal()
    def __gt__(self,other):
        return self.convert_decimal()>other.convert_decimal()
````

````python
obj1 = Fraction(5,8)
obj2 = Fraction(2,3)
````

````python
# str method
print(obj1)
print(obj2)
````

````python
print(obj1 + obj2)
print(obj1 - obj2)
print(obj1 * obj2)
print(obj1 / obj2)
print(obj1.convert_decimal())
print(obj1==obj2)
print(obj1<obj2)
print(obj1>obj2)
````

## WAP to create a class DATA with values as data member. Implement less than, greater than, and equality operator for two objects of the same class

````python
class Data:
    def __init__(self,value):
        self.value = value
    def __eq__(self,other):
        return self.value == other.value
    def __lt__(self,other):
        return self.value < other.value
    def __gt__(self,other):
        return self.value > other.value
````

````python
d1 = Data(30)
d2 = Data(40)
````

````python
print(d1 == d2)
print(d1 < d2)
print(d1 > d2)
````

## METHOD RESOLUTION ORDER IN PYTHON
- It defines an order in which to search for the class to find out which method to use in case of multiple inheritance
- It is done through C3 linearlization algorithm
  

## Example Demonstration

````python
class F:
    pass
class E:
    pass
class D:
    pass
class C(D,F):
    pass
class B(D,E):
    pass
class A(B,C):
    pass
````

````python
A.mro()
````

# First Step

```
L[D] = DO
L[E] = EO
L[F] = FO

# Second Step 
```
MRO OF C = C + MERGE(L[D],L[F],DF)
= C + MERGE(DO,FO,DF)
= C + D + MERGE(O,FO,F)
= C + D + F + O
= CDFO
```

# Third Step
```
MRO OF B = B + MERGE(L[D],L[E],DE)
= B + MERGE(DO,EO,DE)
= B + D + MERGE(O,EO,E)
= B + D + E + O
= BDEO
```

# FORTH Step
```
MRO OF A = A + MERGE(L[B],L[C],BC)
= A + MERGE(BDEO,CDFO,BC)
= A + B + MERGE(DEO,CDFO,C)
= A + B + C + MERGE(DEO,DFO)
= A + B + C + D + MERGE(EO,FO)
= A + B + C + D + E + MERGE(O,FO)
= A + B + C + D + E + F + O
= ABCDEFO
```
